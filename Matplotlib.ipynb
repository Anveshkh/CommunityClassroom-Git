```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import matplotlib
print(matplotlib.__version__)
```

```python
x = np.array([1,6])
y = np.array([0, 250])
plt.plot(x,y)
plt.show()
```

```python
a = np.array([2, 3, 4, 5, 6, 7])
b = np.array([4, 14, 36, 45, 23, 90])
plt.plot(a,b, marker = 'o')
plt.show()
```

```python
a = np.array([1,2,3,4,5])
b = np.array([1,4,9,16,25])
plt.plot(a,b, 'o:r', mfc='b', mec='b')
plt.show()
```

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, linestyle='dotted', marker='o', mec='b', mfc='b')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
x = np.array([80,85,90,100,105,110,115,120,125])
y = np.array([240,250,260,270,290,300,310,320,330])
plt.plot(x,y)
plt.xlabel("Average Pulse", fontdict={"family" : "sans-serif", "color" : "red"})
plt.ylabel("Calorie Burnage", fontdict={"family": "sans-serif", "color" : "red"})
plt.title("Sports Watch Data")
plt.grid()
plt.show()
```
